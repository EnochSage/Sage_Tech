```python
#Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Preparing data
haber_data = pd.read_csv('https://raw.githubusercontent.com/Vihaanshah29/Haberman_dataset_EDA/main/haberman.csv')
```

```python
print(haber_data.head())
```

```python
#Displaying age of cancer patients
haber_data['age']
```

```python
#Displaying year of patient operation
haber_data['year']
```

```python
#Displaying infected lymph node
haber_data['nodes']
```

```python
#Displaying Status of patients. 1 = survived for 5 years or longer, 2 = died within 5 years 
haber_data['status'].value_counts()
```

# Data Analysis and Visualizations

```python
#Scatter Plot
#Nodes against age
haber_data.plot(kind = 'scatter', x = 'age', y = 'nodes')
plt.title('Scatterplot of age vs nodes')
plt.show()
```

```python
#Scatterplot using seaborn
sns.set_style('whitegrid')
sns.FacetGrid(haber_data, hue = 'status', height = 5) \
.map(plt.scatter, 'age', 'nodes')\
.add_legend()
plt.title('Scatterplot using seaborn of age vs nodes')
plt.show()
```

```python
#Pairplot
sns.set_style('whitegrid')
sns.pairplot(haber_data, hue = 'status', height = 3)
plt.show()
```

```python
#Box Plots
sns.boxplot(x = 'status', y = 'nodes', data = haber_data)
plt.legend
plt.title('Box Plot')
plt.show()
```

```python
#Histogram of age
#The line reps the Probability Density Function(PDF)
sns.FacetGrid(haber_data, hue = 'status', height = 8)\
.map(sns.distplot, 'age')\
.add_legend()
plt.title('Histogram of age')
```

```python
#Violin Plot
sns.violinplot(x = 'status', y = 'nodes', data = haber_data)
plt.legend
plt.title('Violin plot')
plt.show()
```
